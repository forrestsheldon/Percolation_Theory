## Introduction

### Basic Idea

Percolation theory deals with the properties of connected clusters in lattices generated by a probabilistic process.  The two most common processes studied are:

* Site percolation - Each site of a lattice is filled with some probability $p$
* Bond percolation - Neighboring sites are connected with probability $p$

This will focus on site percolation although the methods used should generalize.

We say a lattice **percolates** when there is a connected path/cluster from one side of the lattice to another.  The core problem of percolation theory is to determine the value $p_c$ at which an infinite lattice percolates or forms an infinite cluster.  Note the requirement of an infinite lattice.  For many quantities in percolation theory, the values of key quantities only become sharp when we consider an infinite lattice.  As a quick example:

#### Number of Filled Sites

In an infinite lattice, the fraction of filled sites will approach $p$ as the size $L$ of the lattice goes to $\infty$.  When
we generate a finite lattice, we can consider it to be drawn from a binomial distribution.  For a lattice of $M$ sites, the probability for $N$ of them to be filled is
$$P(N, M) = {M \choose N} p^N (1-p)^{M-N}$$
Near the peak and for large $M$, this will approach a gaussian with mean $pM$ and variance $Mp(1-p)$.  To arrive at this, use Stirling's approximation, take the log and drop all constant terms and terms which are 'small' numbers (assuming $M$ and $N$ are 'large' in Schroeder's taxonomy).  Then expand about the peak and renormalize, arriving at, for a lattice with $L^d$ sites,
$$P(N, M) = \frac{1}{\sqrt{2\pi L^dp(1-p)}}e^{\frac{(N - pL^d)^2}{2L^d p(1-p)}}$$
To avoid the dependence on the lattice size and subsequent divergence in the infinite limit, it is convenient to instead consider the fraction of filled sites or occupied sites per site.  Changing variables to $n=N/L^d$ we have,
$$dP = \frac{1}{\sqrt{2\pi L^d p(1-p)}}e^{\frac{(N - pL^d p(1-p))^2}{2L^d p(1-p)}}dN =
\sqrt{\frac{L^d}{2\pi p(1-p)}}e^{\frac{(n - p)^2}{2(p(1-p)/L^d)}}dn$$
Thus the mean is $p$ with a variance of $\frac{p(1-p)}{L^d}$.  We can take this as a parable in that, in the limit of an infinite lattice, quantities will approach their sharp values as $\frac{C}{system size}$.

Other interesting questions we could ask include:

* How many clusters of size $s$ do we find?
* What is the average size of a cluster?
* How big is the infinite cluster as we increase $p$ past $p_c$

As many of these properties are intractable analytically, we will focus on their behavior locally near the critical point, deriving critical exponents that describe how the quantities scale in factors of $(p-p_c)$.  We begin with the 1D case to develop intuition.

## 1D Percolation Theory

```python
def perc_lattice_1D(L, p):
    return np.random.choice([1, 0], (1,L), p=[p, 1-p])

def perc_lattice_2D(L, p):
    return np.random.choice([1, 0], (L,L), p=[p, 1-p])

def display_lattice(lattice):
    plt.matshow(lattice, cmap='binary', interpolation='none')
```

```python
latt = perc_lattice_1D(100, .7)
display_lattice(latt)
```

#### Critical point

In a 1D lattice a single unoccupied site breaks the infinite cluster and thus $p_c=1$.  We can prove this by finding the probability that a finite lattice percolates $\Pi(p,L)$ and taking the limit at $L\to\infty$.

#### Cluster Number Distribution

$n_s(p)$ is the number of clusters of size $s$ per site (or per unit length). The probability of a particular $s$ cluster is $(1-p)^2 p^s$.  We pick a particular site in the cluster.  The expected number of times this site will occur at position $i$ in the lattice is $(1-p)^2 p^s$ and thus the expected number of clusters per site is $n_s(p) = (1-p)^2 p^s$.

#### Cutoff Cluster Size

We can rewrite,
$$n_s(p) = (1-p)^2 \exp{(s\ln{p})} = (1-p)^2 \exp{(-\frac{s}{s_\xi})}$$
with $s_\xi = -\frac{1}{\ln p}$ defined as the **cutoff cluster size/characteristic cluster size**.  We thus see that clusters are exponentially surpressed in the size of the cluster.

This diverges as we approach the critical point. By expanding  we find that
$$s_\xi \propto |p_c - p|^{-1}\quad as \quad p\to p_c.$$
We introduce the **critical exponent $\sigma$,**
$$s_\xi \propto |p_c - p|^{-\frac{1}{\sigma}}\quad as \quad p\to p_c.$$
Thus for 1D percolation, $\sigma=1$.

#### Mean Cluster Size

The probability that an arbitrary site is part of an $s$ cluster is $sn_s(p)$.  This is also the fraction of all sites that are in $s$ clusters.  As the fraction of sites in *any* cluster is $p$ we have,
$$\sum_s sn_s(p) = p$$
which will be a useful identity later.

The probability that an occupied site is part of an $s$ cluster is $\frac{n_s(p)}{p} = \frac{sn_s(p)}{\sum s n_s(p)}$.
